```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. Загружаем датасет
df = pd.read_csv('rnc_dataset_markup.csv', sep=';')

# 2. Стратифицированное разбиение на train (80%) и test (20%)
#    Целевой столбец: shouldSplit
train_idx, test_idx = train_test_split(
    df.index,
    test_size=0.2,
    random_state=42,
    stratify=df['shouldSplit']      # стратификация по shouldSplit
)

# 3. Добавляем столбец split
df['split'] = 'train'
df.loc[test_idx, 'split'] = 'test'

# 4. Проверка (покажет, что пропорции shouldSplit в train и test близки)
print("Train class distribution:")
print(df[df['split']=='train']['shouldSplit'].value_counts(normalize=True))
print("\nTest class distribution:")
print(df[df['split']=='test']['shouldSplit'].value_counts(normalize=True))

# 5. Сохраняем
df.to_csv('rnc_dataset_markup_with_split.csv', index=False)
```
