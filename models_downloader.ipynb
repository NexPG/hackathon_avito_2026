```python
import os
import zipfile
import gdown
from tqdm.auto import tqdm

MODEL_ZIPS = {
    "final_detected_model": "1EpFmlmcaMQS19cHbbJv1tRyQhbHfdQzH",
    "final_split_model":    "1yy-E_XewCEfumTu7dZfVkh4A8b88K7um",
}

def download_and_extract(model_dir, file_id):
    url = f"https://drive.google.com/uc?id={file_id}"
    zip_path = f"{model_dir}.zip"
    
    print(f"📦 Папка '{model_dir}' не найдена. Скачиваю архив...")
    try:
        gdown.download(url, zip_path, quiet=False)
    except Exception as e:
        print(f"❌ Ошибка скачивания: {e}")
        print("⚠️ Google Диск временно недоступен или ссылка недействительна.")
        print(f"👉 Пожалуйста, обучите модель '{model_dir}' заново, запустив соответствующую ячейку.")
        raise
    
    print(f"📂 Распаковываю {zip_path}...")
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        members = zip_ref.infolist()
        with tqdm(total=len(members), desc=f"Распаковка {model_dir}", unit="файл") as pbar:
            for member in members:
                zip_ref.extract(member, path=".")
                pbar.update(1)  # ← исправлено: используем pbar, а не tqdm
    os.remove(zip_path)
    print(f"✅ Модель '{model_dir}' готова.\n")

def check_and_download_models():
    missing = [d for d in MODEL_ZIPS if not os.path.isdir(d)]
    if not missing:
        print("✅ Все необходимые модели уже присутствуют в директории.")
        return
    
    print(f"⚠️ Отсутствуют папки: {', '.join(missing)}")
    print("🔄 Будет выполнено скачивание с Google Диска...")
    for model_dir in missing:
        download_and_extract(model_dir, MODEL_ZIPS[model_dir])
    print("🎉 Готово! Модели загружены и распакованы.")

check_and_download_models()
```
