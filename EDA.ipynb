```python
import pandas as pd
import ast
from datasets import Dataset
```

```python
df = pd.read_csv('rnc_dataset.csv')
df['targetDetectedMcIds'] = df['targetDetectedMcIds'].apply(ast.literal_eval)
df['targetSplitMcIds'] = df['targetSplitMcIds'].apply(ast.literal_eval)

# 1. Проверка уникальных категорий и их частоты
all_labels = sorted(set([label for labels in df['targetDetectedMcIds'] for label in labels]))
print(f"Всего категорий: {len(all_labels)}")

# 2. Разделение данных по столбцу split
train_df = df[df['split'] == 'train']
val_df = df[df['split'] == 'val']
test_df = df[df['split'] == 'test']

print(train_df)
print(val_df)
print(test_df)
```

```python
def prepare_labels(df):
    label_matrix = pd.DataFrame(0, index=range(len(df)), columns=all_labels)
    for idx, labels in enumerate(df['targetDetectedMcIds']):
        for label in labels:
            if label in label_matrix.columns:
                label_matrix.iloc[idx, label_matrix.columns.get_loc(label)] = 1
    return label_matrix

# Создаем метки для каждого набора данных
train_labels = prepare_labels(train_df)
val_labels = prepare_labels(val_df)
test_labels = prepare_labels(test_df)

print(train_labels)
print(val_labels)
print(test_labels)
```
